```python
import numpy as np
import pandas as pd
import requests
import seaborn as sns
import matplotlib.pyplot as plt
import csv
from bs4 import BeautifulSoup
```

```python
url = 'https://www.iplt20.com/points-table/men/2020'
```

```python
r = requests.get(url)
```

```python
content = BeautifulSoup(r.content, 'html.parser')
```

```python
ipl = content.find_all('table', class_ = 'standings-table standings-table--full')
```

```python
ipl
```

```python
Team = []
Played = []
Won = []
Lost = []
Tied = []
NR = []
For = []
Against = []
Points = []
```

```python
for data in ipl:   
    team = data.a.find('span', attrs = {'class':'standings-table__team-name js-team'}).text
    Team.append(team)
    
    play = data.find('td', attrs = {'class':'standings-table__padded'}).text
    Played.append(play)
    
    value = data.find_all('td', attrs = {'class':'standings-table__optional'})
    won = value[0].text
    Won.append(won)
    
    lost = value[1].text
    Lost.append(lost)
    
    tied = value[2].text
    Tied.append(tied)
    
    nr = value[3].text
    NR.append(nr)
    
    fr = value[4].text
    For.append(fr)
    
    against = value[5].text
    Against.append(against)
    
    points = data.find('td', attrs = {'class':'standings-table__highlight js-points'}).text
    Points.append(points)
```

```python
Points
```

```python
lets = pd.DataFrame({'Team':Team, 'Played':Played, 'Won':Won, 'Lost':Lost, 'Tied':Tied, 'Not Result':NR, 'For':fr, 'Against':Against, 'Points':Points})
```

```python
lets
```

```python
lets.to_csv('ipl.csv')
```
